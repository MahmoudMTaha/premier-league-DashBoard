# Feature Engineering

```python
df['total_goals']=df['FTHG']+df['FTAG']
```

```python
df2=pd.DataFrame(df.groupby(df.Season)['total_goals'].sum())
df2=df2.reset_index()
df2
```

```python
df3=df.copy()
df3=pd.DataFrame(df3[df['FTR'] =='H'].groupby([df.Season,'HomeTeam']).agg({'total_goals':'sum','FTR':'count'}))
df3=df3.reset_index()
df3[df3['Season']==2018]
```

```python
df3['HomeTeam'].unique()
```

```python
df4=df.copy()
home_wins=df4[df['FTR'] =='H'].groupby([df.Season,'HomeTeam']).agg({'FTR':'count','FTHG':'sum'}).reset_index()
home_wins.rename(columns={'FTR':'HomeWins'},inplace=True)
home_draw=df4[df['FTR'] =='D'].groupby([df.Season,'HomeTeam']).agg({'FTR':'count','FTHG':'sum'}).reset_index()
home_draw.rename(columns={'FTR':'HomeDraw'},inplace=True)
away_wins=df4[df['FTR'] =='A'].groupby([df.Season,'AwayTeam']).agg({'FTR':'count','FTAG':'sum'}).reset_index()
away_wins.rename(columns={'FTR':'AwayWins'},inplace=True)
away_draw=df4[df['FTR'] =='D'].groupby([df.Season,'AwayTeam']).agg({'FTR':'count','FTAG':'sum'}).reset_index()
away_draw.rename(columns={'FTR':'AwayDraw'},inplace=True)
away_draw[away_draw['AwayTeam']=='Chelsea']
```

```python
df5=pd.merge(left=home_wins,right=away_wins,left_on=['Season','HomeTeam'],right_on=['Season','AwayTeam'])
df5.drop('AwayTeam',axis=1,inplace=True)
df5.rename(columns={'HomeTeam':'Team'},inplace=True)
df5[df5['Season']==2018]
```

```python
df6=pd.merge(left=home_draw,right=away_draw,left_on=['Season','HomeTeam'],right_on=['Season','AwayTeam'],how='right')

df6.drop('AwayTeam',axis=1,inplace=True)
df6.rename(columns={'HomeTeam':'Team'},inplace=True)
df6.loc[123,['Team','HomeDraw','FTHG']]=['Chelsea',0,0]
df6.loc[47,['Team','HomeDraw','FTHG']]=['Man United',0,0]
df6.loc[172,['Team','HomeDraw','FTHG']]=['Man City',0,0]
df6[df6['Season']==2018]
```

```python
df7=pd.merge(left=df5,right=df6,left_on=['Season','Team'],right_on=['Season','Team'])
```

```python
df7['TotalWins']=df7['HomeWins']+df7['AwayWins']
df7['TotalDraw']=df7['HomeDraw']+df7['AwayDraw']
df7['TotalPoints'] = 3*df7['TotalWins'] + df7['TotalDraw']
df7['TeamGoals']=df7['FTHG_x']+df7['FTAG_x']+df7['FTHG_y']+df7['FTAG_y']
```

```python
df7['TeamRank']=df7.groupby('Season')['TotalPoints'].rank(method='min',ascending=False)
```

```python
df7=df7[df7['TeamRank']==1]
df7=df7.drop_duplicates(subset=['Season'],keep='first')
df7
```

```python
import plotly.graph_objects as go

@app.callback(
    Output(component_id='my_Ddiv',component_property='figure'),
    Output(component_id='my_Ddiv_new',component_property='figure'),
    Output(component_id='team_name',component_property='children'),
    Output(component_id='Matches',component_property='children'),
    Output(component_id='season',component_property='children'),
    Output(component_id='win_rate',component_property='children'),
    Output(component_id='goals_rate',component_property='children'),
    Input(component_id='Arsenal',component_property='n_clicks'),
    Input(component_id='Aston Villa',component_property='n_clicks'),
    Input(component_id='Birmingham',component_property='n_clicks'),
    Input(component_id='Blackburn',component_property='n_clicks'),
    Input(component_id='Blackpool',component_property='n_clicks'),
    Input(component_id='Bolton',component_property='n_clicks'),
    Input(component_id='Chelsea',component_property='n_clicks'),
    Input(component_id='Everton',component_property='n_clicks'),
    Input(component_id='Fulham',component_property='n_clicks'),
    Input(component_id='Liverpool',component_property='n_clicks'),
    Input(component_id='Man City',component_property='n_clicks'),
    Input(component_id='Man United',component_property='n_clicks'),
    Input(component_id='Newcastle',component_property='n_clicks'),
    Input(component_id='Stoke',component_property='n_clicks'),
    Input(component_id='Sunderland',component_property='n_clicks'),
    Input(component_id='Tottenham',component_property='n_clicks'),
    Input(component_id='West Brom',component_property='n_clicks'),
    Input(component_id='West Ham',component_property='n_clicks'),
    Input(component_id='Wigan',component_property='n_clicks'),
    Input(component_id='Wolves',component_property='n_clicks'),
    Input(component_id='Norwich',component_property='n_clicks'),
    Input(component_id='QPR',component_property='n_clicks'),
    Input(component_id='Swansea',component_property='n_clicks'),
    Input(component_id='Reading',component_property='n_clicks'),
    Input(component_id='Southampton',component_property='n_clicks'),
    Input(component_id='Cardiff',component_property='n_clicks'),
    Input(component_id='Crystal Palace',component_property='n_clicks'),
    Input(component_id='Hull',component_property='n_clicks'),
    Input(component_id='Burnley',component_property='n_clicks'),
    Input(component_id='Leicester',component_property='n_clicks'),
    Input(component_id='Bournemouth',component_property='n_clicks'),
    Input(component_id='Watford',component_property='n_clicks'),
    Input(component_id='Middlesbrough',component_property='n_clicks'),
    Input(component_id='Brighton',component_property='n_clicks'),
    Input(component_id='Huddersfield',component_property='n_clicks'),
    Input(component_id='Sheffield United',component_property='n_clicks'),
    Input(component_id='Leeds',component_property='n_clicks'),
)
def update_My_Div(Arsenal,AstonVilla,Birmingham,Blackburn,Blackpool,
                  Bolton,Chelsea,Everton,Fulham,Liverpool,ManCity,ManUnited,
                  Newcastle,Stoke,Sunderland,Tottenham,WestBrom,WestHam,Wigan,
                  Wolves,Norwich,QPR,Swansea,Reading,Southampton,Cardiff,CrystalPalace,
                  Hull,Burnley,Leicester,Bournemouth,Watford,Middlesbrough,Brighton,
                  Huddersfield,SheffieldUnited,Leeds):
    ctx = dash.callback_context
    if not ctx.triggered:
        input_value = 'Arsenal'
    else:
        input_value = ctx.triggered[0]['prop_id'].split('.')[0]
    number_of_matches = len(df8[(df8['HomeTeam'] == input_value) | (df8['AwayTeam'] == input_value)])
    seasons = number_of_matches / 38
    win_h = df8[(df8['HomeTeam'] == input_value) & (df8['FTR'] == 'H') ]['HomeTeam'].count()
    draw_h = df8[(df8['HomeTeam'] == input_value) & (df8['FTR'] == 'D') ]['HomeTeam'].count()
    loss_h = df8[(df8['HomeTeam'] == input_value) & (df8['FTR'] == 'A') ]['HomeTeam'].count()
    loss_a = df8[(df8['AwayTeam'] == input_value) & (df8['FTR'] == 'H') ]['AwayTeam'].count()
    draw_a = df8[(df8['AwayTeam'] == input_value) & (df8['FTR'] == 'D') ]['AwayTeam'].count()
    win_a = df8[(df8['AwayTeam'] == input_value) & (df8['FTR'] == 'A') ]['AwayTeam'].count()
    wins_rate = round(((win_h + win_a) / number_of_matches) * 100,2)
    fig = go.Figure()
    fig.add_trace(go.Bar(
        y = ['Loss', 'Draw', 'Win'],
        x = [loss_a, draw_a, win_a],
        name = 'Away',
        orientation='h',
        marker=dict(
            color='rgba(243, 32, 32, 0.6)',
            line=dict(color='rgba(243, 32, 32, 1.0)', width=3)
        )
    ))
    fig.add_trace(go.Bar(
        y = ['Loss', 'Draw', 'Win'],
        x = [loss_h, draw_h, win_h],
        name = 'Home',
        orientation = 'h',
        marker = dict(
            color='rgba(12, 255, 68, 0.6)',
            line=dict(color='rgba(12, 255, 68, 1.0)', width=3)
        )
    ))
    fig.update_layout(barmode='stack')
#     fig =go.Figure(go.Sunburst(
#         labels=[input_value, "Home","Win","Draw","Loss","Away","Win.","Draw.","Loss."],
#         parents=["", input_value, "Home", "Home", "Home",input_value,"Away","Away","Away"],
#         values=[win_h+draw_h+loss_h+loss_a+draw_a+win_a, win_h+draw_h+loss_h, win_h, draw_h, loss_h,
#                 win_a+draw_a+loss_a, win_a, draw_a, loss_a],
#     ))
    
    home_corners = df8[(df8['HomeTeam'] == input_value)]['HC'].sum()
    home_fouls_committed = df8[(df8['HomeTeam'] == input_value)]['HF'].sum()
    home_yellow_cards = df8[(df8['HomeTeam'] == input_value)]['HY'].sum()
    home_red_cards = df8[(df8['HomeTeam'] == input_value)]['HR'].sum()
    away_corners = df8[(df8['AwayTeam'] == input_value)]['AC'].sum()
    away_fouls_committed = df8[(df8['AwayTeam'] == input_value)]['AF'].sum()
    away_yellow_cards = df8[(df8['AwayTeam'] == input_value)]['AY'].sum()
    away_red_cards = df8[(df8['AwayTeam'] == input_value)]['AR'].sum()
    home_total_shots = df8[(df8['HomeTeam'] == input_value)]['HS'].sum()
    home_target_shots = df8[(df8['HomeTeam'] == input_value)]['HST'].sum()
    home_goals = df8[(df8['HomeTeam'] == input_value)]['FTHG'].sum()
    away_total_shots = df8[(df8['AwayTeam'] == input_value)]['HS'].sum()
    away_target_shots = df8[(df8['AwayTeam'] == input_value)]['HST'].sum()
    away_goals = df8[(df8['AwayTeam'] == input_value)]['FTHG'].sum()
    goals_rate = round((home_goals + away_goals) / number_of_matches,2)


    Home_stat = (home_corners + home_fouls_committed + home_yellow_cards + home_red_cards)
    Away_stat = (away_corners + away_fouls_committed + away_yellow_cards + away_red_cards)
    statistics = Home_stat + Away_stat
    Attacking = home_total_shots + away_total_shots
    summary = Attacking + statistics
             
    fig2 = go.Figure(go.Sunburst(
        ids = [input_value, 'statistics','Attacking','Home_stat','Away_stat','Home_attack','Away_attack',
               'Corners_Home','Fouls_Committed_Home','Yellow_cards_Home','Red_Cards_Home',
               'Corners_Away','Fouls_Committed_Away','Yellow_cards_Away','Red_Cards_Away',
               'home_total_shots','home_target_shots','home_goals',
               'away_total_shots' , 'away_target_shots' , 'away_goals'
               ],
        labels = [input_value,'statistics','Attacking','Home','Away','Home_shots','Away_shots',
                  'Corners','Fouls<br>Committed','Yellow<br>cards','Red<br>Cards',
                  'Corners','Fouls<br>Committed','Yellow<br>cards','Red<br>Cards',
                  'Shots on Target','Goals',
                  'Shots on Target','Goals'
                 ],

        parents=["", input_value, input_value, "statistics","statistics",'Attacking','Attacking',
                 "Home_stat",'Home_stat',"Home_stat","Home_stat",
                 "Away_stat",'Away_stat','Away_stat','Away_stat',
                 'Home_attack','Home_attack',
                 'Away_attack','Away_attack'
                ],

        values=[summary,statistics,Attacking,Home_stat,Away_stat,home_total_shots,away_total_shots,
                home_corners , home_fouls_committed , home_yellow_cards , home_red_cards ,
               away_corners , away_fouls_committed , away_yellow_cards , away_red_cards,
                home_target_shots,home_goals,
                away_target_shots,away_goals
               ],
    ))
    return fig,fig2,input_value,number_of_matches,seasons,f'{wins_rate} %',goals_rate
```
